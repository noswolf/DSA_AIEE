## 📌 Quick Guide: Using `input()`

In Python, `input()` is how your program asks the user to type something while the program is running.

```python
name = input()      # waits for the user to type something and press Enter
print(name)          # prints whatever they typed
```

A few things to know:

- **`input()` always returns a string.** Even if the user types a number like `25`, Python stores it as the text `"25"`, not the integer `25`.
- **You can add a prompt message** so the user knows what to type, e.g. `input("Enter your name: ")`. In this exercise, no prompt text is shown — the program just silently waits for each line of input, in the order given by the test case (first the name, then the age).
- **If you need a number**, convert the string using `int()` or `float()`:
```python
  age = input()
  age = int(age)   # now age is a number, not text
```
- **Each call to `input()` reads one line.** Since this problem gives two lines of input (name, then age), you'll call `input()` twice:
```python
  name = input()
  age = input()
```
