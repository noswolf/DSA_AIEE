## Lab 6-3: Write a insertionSort on an input list

Insertion Sort works the way many people naturally organize a hand of playing cards. It divides the list into a "sorted sub-list" (which starts at the front) and an "unsorted sub-list." Each round, it takes the first element from the unsorted section and "inserts" it into its correct, sorted position by shifting larger elements out of the way.

### Hint

Nested Loop is required (i.e. outer and inner loop)
- A while loop is required as inner loop

1. The Outer Loop (`index`): Unlike previous algorithms, you start this loop at index `1` (the second element) instead of `0`. Why? Because a list containing just one element (`alist[0]`) is already technically sorted!
2. The Target Elements (`currentvalue` and `position`): At the beginning of each round, copy the value you want to insert into a temporary variable called `currentvalue`. You will also need a tracker variable, `position`, to keep tabs on where this value might end up as you slide down the sorted section.
3. The Shifting Mechanism (The `while` Loop): Instead of using full, multi-step swaps like Bubble or Selection Sort, Insertion Sort *shifts* elements.
   * The `while` loop needs to look backward through the sorted section (to the left). It keeps running as long as `position > 0` and the element to its left (`alist[position-1]`) is greater than your `currentvalue`.
   * Inside this loop, you copy the larger element one slot to the right (`alist[position] = alist[position-1]`) and move your `position` tracker one index to the left.
4. Tracking Metrics:
   * **Exchanges:** Every time you successfully shift an element to the right inside the `while` loop, increment your `no_exchange` counter.
   * **Comparisons:** You must count the comparisons where `currentvalue < alist[position-1]` is evaluated.
      * *Crucial Note:* Remember that a `while` loop condition eventually evaluates to `False` to stop the loop. If the loop stops because `currentvalue` was *not* smaller than `alist[position-1]` (and not because it ran out of room at index `0`), that final check still counts as a comparison! You will need an extra check after the `while` loop to account for this.
5. **The Final Insertion**: Once the `while` loop finishes shifting elements, the `position` tracker will be sitting exactly where `currentvalue` belongs. Drop `currentvalue` into `alist[position]`.
6. Printing Requirements:
   * Print the round number and the current value you are trying to insert at the beginning of each outer loop iteration.
   * Print the final position to insert and the current state of the list after the value has been placed.
   * Print the total `no_compare` and `no_exchange` metrics at the very end of the function.
