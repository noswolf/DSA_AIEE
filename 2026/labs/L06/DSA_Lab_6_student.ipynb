```python
from io import StringIO
import sys

def test_bubbleSort():
    """
    Tests the bubbleSort function using a variety of test cases.
    """
    test_cases = [
        {
            "name": "best_case_scenario",
            "input": [1, 2, 3, 4, 5],
            "expected_output": [1, 2, 3, 4, 5],
            "expected_comparisons": 10,
            "expected_exchanges": 0
        },
        {
            "name": "worst_case_scenario",
            "input": [5, 4, 3, 2, 1],
            "expected_output": [1, 2, 3, 4, 5],
            "expected_comparisons": 10,
            "expected_exchanges": 10
        },
        {
            "name": "random_list_1",
            "input": [3, 1, 4, 1, 5, 9, 2, 6],
            "expected_output": [1, 1, 2, 3, 4, 5, 6, 9],
            "expected_comparisons": 28,
            "expected_exchanges": 8
        },
        {
            "name": "empty_list",
            "input": [],
            "expected_output": [],
            "expected_comparisons": 0,
            "expected_exchanges": 0
        },
        {
            "name": "single_element",
            "input": [7],
            "expected_output": [7],
            "expected_comparisons": 0,
            "expected_exchanges": 0
        },
        {
            "name": "duplicate_elements",
            "input": [4, 2, 4, 1, 2],
            "expected_output": [1, 2, 2, 4, 4],
            "expected_comparisons": 10,
            "expected_exchanges": 6
        },
        {
            "name": "negative_numbers",
            "input": [-3, 0, -1, 2, -5],
            "expected_output": [-5, -3, -1, 0, 2],
            "expected_comparisons": 10,
            "expected_exchanges": 5
        }
    ]

    passed_cases = 0
    failed_cases = 0
    
    print("Running bubbleSort tests...")

    for test_case in test_cases:
        # Create a copy of the input list to avoid modifying the original test case
        input_list = test_case["input"].copy()

        # Redirect the print output to not show the print statements in bubblesort in the console
        old_stdout = sys.stdout
        sys.stdout = captured_output = StringIO()

        # Run bubble sort and capture the output, comparisons, and exchanges
        bubbleSort(input_list)

        # Restore stdout
        sys.stdout = old_stdout

        output_lines = captured_output.getvalue().strip().split('\n')
        
        # Extract the number of comparisons and exchanges from the output
        comparisons_line = [line for line in output_lines if "Number of Comparisons" in line][0]
        exchanges_line = [line for line in output_lines if "Number of Exchanges" in line][0]
        
        actual_comparisons = int(comparisons_line.split(":")[1].strip())
        actual_exchanges = int(exchanges_line.split(":")[1].strip())
        
        if (input_list == test_case["expected_output"] and
                        actual_comparisons == test_case["expected_comparisons"] and
                        actual_exchanges == test_case["expected_exchanges"]):
            print(f"Test case '{test_case['name']}' passed!")
            passed_cases += 1
        else:
            print(f"Test case '{test_case['name']}' failed:")
            if input_list != test_case["expected_output"]:
                print(
                    f"\tExpected output {test_case['expected_output']}, but got {input_list}")
            if actual_comparisons != test_case["expected_comparisons"]:
                print(
                    f"\tExpected {test_case['expected_comparisons']} comparisons, but got {actual_comparisons}")
            if actual_exchanges != test_case["expected_exchanges"]:
                print(
                    f"\tExpected {test_case['expected_exchanges']} exchanges, but got {actual_exchanges}")
            failed_cases += 1
    
    print(f"\nSummary:")
    print(f"Passed cases: {passed_cases}")
    print(f"Failed cases: {failed_cases}")
```

```python
# Call the test function
test_bubbleSort()
```
