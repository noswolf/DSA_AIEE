### while loop

```python
i = 0   # Initialize variable first
while i < 5: # write your own condition here
  print(i)
  i += 1 # Equivalent to i = i + 1
print("Loop finished")
```

```python
lst = ["apple", 10, "banana"]
i = 0
how_many = len(lst)

while i < how_many: # write your own condition here
  print(lst[i])
  i += 1 # Equivalent to i = i + 1
```

## 2. Loop through a string

```python
item = "DSA_AIEE_2026"
```

### For loop

```python
for i in item:
    print(i)
```

### while loop

```python
item = "DSA_AIEE_2026"
len(item)
```

```python
i = 0
while i < len(item) and i==0: # write your own condition here
  print(item[i])
  i += 1    # Equivalent to i = i + 1
print("Run finished")
```

# Mod Operator (%)

Mod is a sibling to division operator (%) where it returns the **remainder**, not the quotient.

### Examples

```python
4/2 # 4 divided by 2
```

```python
4%2 # 4 mod by 2
```

```python
7/3
```

```python
7%3
```

### Example: while loop and print only even number

```python
i = 0
while i < 10: # write your own condition here
  if i % 2 == 0:
    print(f"{i} is an even number")
  i += 1
```

### Example: while loop and print only divisible number

```python
i = 0
while i < 10: # write your own condition here
  if i % 2 == 0:
    print(f"{i} is divisible by 2")
  if i % 3 == 0:
    print(f"{i} is divisible by 3")
  else:
    print(f"{i} is not divisible by 2 or 3")
  i += 1
```
