- If all tests pass, you'll see 'All test cases passed' (which is good!).
- If a test fails, you'll see an AssertionError indicating which test case failed and why.

```python
def test_find_peak_element():
    test_cases = [
        {'input': [1, 3, 2, 1], 'expected': 1},
        {'input': [1, 2, 3, 1], 'expected': 2},
        {'input': [3, 2, 1], 'expected': 0},
        {'input': [1, 2, 1, 3, 5, 6, 4], 'expected': 5},
        {'input': [1], 'expected': 0},
        {'input': [1, 2], 'expected': 1},
        {'input': [2, 1], 'expected': 0},
        {'input': [], 'expected': -1},
        {'input': 5, 'expected': -1}
    ]

    for case in test_cases:
        arr = case['input']
        expected = case['expected']
        result = findPeakElement(arr)
        # For find_peak_element, we need to check if the returned index is a valid peak
        if result != -1:
            is_peak = True
            if result > 0 and arr[result] <= arr[result - 1]:
                is_peak = False
            if result < len(arr) - 1 and arr[result] <= arr[result + 1]:
                is_peak = False
            assert is_peak, f"For input {arr}: {result} is not a peak"
        else:
            assert result == expected, f"For input {arr}: expected {expected}, but got {result}"
            
    print("\n#############################################")
    print("All test cases passed!")
    print("#############################################")

# Run the tests
if __name__ == "__main__":
    test_find_peak_element()
```
