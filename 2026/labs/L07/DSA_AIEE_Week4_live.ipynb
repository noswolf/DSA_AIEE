```python
class BankAccount:
    country = "Thailand"
    def __init__(self, owner, balance):
        self.owner = owner          # 'self.owner' stores data ON this specific object
        self.balance = balance

    def show_info(self):
        print(f"Account of {self.owner}, Balance: {self.balance}")

    def abcdef(self):
        self.balance = 0
        self.owner = self.owner
        self.country = "USA"

# Creating two different objects from the same class
acc1 = BankAccount("Alice", 5000)
acc2 = BankAccount("Bob", 1200)

acc1.show_info()  # Output: Account of Alice, Balance: 5000
acc2.show_info()  # Output: Account of Bob, Balance: 1200
```

# 2. setter/getter

```python
class BankAccount:
    country = "Thailand"
    def __init__(self, owner, balance):
        self._owner = owner          # 'self.owner' stores data ON this specific object
        self._balance = balance

    def show_info(self):
        print(f"Account of {self._owner}, Balance: {self._balance}")

    def abcdef(self):
        self._balance = 0
        self._owner = self.owner
        self.country = "USA"

    # Getter method for owner attribute
    def get_owner(self):
        return self._owner

    # Setter method for owner attribute
    def set_owner(self, new_owner):
        self._owner = new_owner

# Creating two different objects from the same class
acc1 = BankAccount("Alice", 5000)
acc2 = BankAccount("Bob", 1200)

acc1.show_info()  # Output: Account of Alice, Balance: 5000
acc2.show_info()  # Output: Account of Bob, Balance: 1200
```

```python
# Direct way to access the attribute (we shouldn't access this way)

acc1._owner
```

```python
# This is how to access the attribute provided by the developer

acc1.get_owner()
```

```python
acc1.set_owner("Alicia")
```

```python
acc1.get_owner()
```
