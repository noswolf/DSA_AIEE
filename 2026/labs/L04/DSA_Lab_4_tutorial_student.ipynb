When a method is called, a special parameter (`self`) identify the particular instance upon which a member is invoked.

```python
class Foo:
    def bar(self):
        self.x = 'Some value'
        print('Bar called, self.x is', self.x)
```

## Abstract Base Class

```python
class Shape:
    def __init__(self, name):
        self.name = name

    def __repr__(self):
        return self.name

    def __str__(self):
        return self.name.upper() #uppercase

    def area(self):
        raise NotImplementedError()
```

### `__init__` method (initialize the attributes/variables of an object)

Create an instance of the Shape class by calling `__init__` method that serves as the **constructor** of the class

`__init__` method establish the state of a newly created shape object with instance variables (in this case, `name` variable)

```python
s = Shape('circle')
```

### `__repr__` method (object representation)

- Provides the "official" string representation of an object (used by `repr(obj)` and the interactive REPL).
- Differs from `__str__`: `__str__` is for a user-friendly display, `__repr__` is for developer-facing representation.

```python
s
```

```python
repr(s)
```

```python
s.name
```

### `__str__` method ()

- A special ("dunder") method that defines the **human-readable, user-friendly string representation** of a class object. It is automatically called when you use functions like `print()`, `str()`, and `format()` on an instance of the class. 
- However, we defined it to return the uppercase name, so `str(s)` or `print(s)` will display 'CIRCLE'.

```python
str(s)
```

```python
s.area()
```

## Inheritance

Create a `Circle` class based on `Shape` class

### `super()`

- `super()` returns a proxy object that delegates method calls to a **superclass** following the method resolution order (MRO).  
- Common use: call a parent initializer or method (e.g., `super().__init__(...)`) so a subclass reuses base-class behavior.  
- Enables cooperative multiple inheritance and avoids hard-coding base class names.  
- In Python 3 you can call `super()` without arguments inside instance methods.

```python
class Circle(Shape):
    def __init__(self, radius):
        super().__init__('circle')
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2
```

```python
c = Circle(5.0)
c
```

```python
c.name
```

```python
c.area()
```
